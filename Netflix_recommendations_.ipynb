```python
import pandas as pd
```

```python
movie=pd.read_csv("/content/Netflix_Dataset_Movie.csv")
movie.head()
```

```python
ratings=pd.read_csv("/content/Netflix_Dataset_Rating.csv.zmdownload")
ratings.head()

```

```python
movie_ratings=pd.merge(movie,ratings)
movie_ratings.head()
```

```python
movie_ratings=movie_ratings.pivot_table(index="User_ID",columns="Name",values="Rating")
movie_ratings.head()
```

```python
movie['Name'][movie['Name'].str.contains('Wild')]
```

```python
wild_rating=movie_ratings['Wild Things']
wild_rating.head()
```

```python
similar_movies=movie_ratings.corrwith(wild_rating)
similar_movies.head()
```

```python
similar_movies=similar_movies.sort_values(ascending=False)
similar_movies.head()s
```
